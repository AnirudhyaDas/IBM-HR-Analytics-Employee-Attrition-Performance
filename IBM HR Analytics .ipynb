# INTRODUCTION

Employee attrition, also known as turnover, is a significant concern for organizations across various industries. High attrition rates can disrupt business operations, lead to increased recruitment and training costs, and result in the loss of valuable knowledge and talent. Understanding the factors that contribute to employee attrition is crucial for organizations to devise strategies that improve retention and reduce associated costs.

In this analysis, we aim to use clustering techniques like K-Means and Hierarchical Clustering to uncover hidden patterns that may lead to employee attrition. By examining key features such as age, job role, and monthly income, we seek to identify distinct employee groups that might have higher tendencies for attrition. This exploratory analysis will provide deeper insights into the underlying employee behavior and allow organizations to design more targeted retention strategies.

## Feature Engineering

```python
df = df.copy()

# 1. Binarize categorical variables using one-hot encoding
df = pd.get_dummies(df, columns=['Department', 'Gender', 'MaritalStatus', 'JobRole'], drop_first=True)

# 2. Create interaction features
# Interaction between MonthlyIncome and JobLevel
df['Income_JobLevel_Interaction'] = df['MonthlyIncome'] * df['JobLevel']

# Interaction between Age and YearsAtCompany
df['Age_YearsAtCompany_Interaction'] = df['Age'] * df['YearsAtCompany']

# 3. Bucket continuous variables
# Bucket Age into groups (young, mid-age, senior)
bins = [18, 30, 40, 50, 60]
labels = ['18-30', '30-40', '40-50', '50-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)

# Bucket MonthlyIncome into quartiles
df['IncomeGroup'] = pd.qcut(df['MonthlyIncome'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])

# 4. Create ratio features
# Ratio of YearsSinceLastPromotion to YearsAtCompany
df['Promotion_Company_Ratio'] = df['YearsSinceLastPromotion'] / (df['YearsAtCompany'] + 1)  # add 1 to avoid division by zero

# 5. Create derived features
# Tenure: Total working experience in the company
df['Tenure'] = df['YearsAtCompany'] + df['YearsInCurrentRole']

# Display the first few rows of the newly engineered dataset
df.head()
```

## Machine Learning Models

#### Data Preparation

```python
X = df.drop(columns=['Attrition', 'AgeGroup', 'IncomeGroup'])
y = df['Attrition'].map({'Yes': 1, 'No': 0})  # Convert 'Attrition' to binary labels
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

categorical_cols = X.select_dtypes(include=['object']).columns

X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42, stratify=y)
```

#### Model Selection and Training

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
```

```python
y_lr_pred= lr_model.predict(X_test)
y_lr_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix
lr_cr=classification_report(y_test, y_lr_pred)
print(lr_cr)
```

```python
# Confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_lr_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

## Hyperparameter Tunning

```python
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'solver': ['liblinear', 'saga']
}

param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}
```

```python
from sklearn.model_selection import GridSearchCV

def perform_grid_search(model, param_grid):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_, grid_search.best_params_
```

```python
# Logistic Regression
best_lr_model, best_lr_params = perform_grid_search(lr_model, param_grid_lr)
```

```python
# Random Forest
best_rf_model, best_rf_params = perform_grid_search(rf_model, param_grid_rf)
```

```python
# XGboost
best_xgb_model, best_xgb_params = perform_grid_search(xgb_model, param_grid_xgb)
```

```python
models = {
    "Logistic Regression": best_lr_model,
    "Random Forest": best_rf_model,
    "XGboost": best_xgb_model
}

results = {}
for model_name, model in models.items():
    y_pred = model.predict(X_test)
    results[model_name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Classification Report": classification_report(y_test, y_pred)
    }
```

## Clustering (Unsupervised Learning)

### K-Means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_encoded)
    sse.append(kmeans.inertia_) 


plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), sse, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(X_encoded)

X_encoded['KMeans_Cluster'] = kmeans_labels

sns.pairplot(X_encoded , hue='KMeans_Cluster', vars=['Age', 'MonthlyIncome'])
plt.show()
```

# CONCLUSION

Through the application of both K-Means and Hierarchical Clustering, we were able to group employees into distinct clusters based on attributes like age, job role, and monthly income. These clusters revealed some hidden patterns that may contribute to employee attrition. For instance, younger employees in certain job roles with lower income levels may exhibit a higher propensity for attrition. Additionally, the hierarchical clustering provided a hierarchical structure of relationships, allowing us to see how different employee attributes merged to form larger groups.

The findings from this analysis can help organizations identify at-risk employee groups and design tailored retention strategies, such as offering competitive compensation packages, providing career development opportunities, and improving work-life balance. By acting on these insights, companies can reduce turnover rates, lower recruitment and training costs, and create a more stable workforce.
